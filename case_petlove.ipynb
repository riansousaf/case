# Importei o pacote pandas para manipular um dataframe.

```python
import pandas as pd
```

# Criei a variável url para armazenar o caminho até a arquivo e a variável petlove_df para armazenar o DataFrame com os dados do arquivo csv.

# Usei a função pd.read_csv() para ler e armazenar os dados e utilizei o parâmetro sep para definir o separador do arquivo.

# Usei a função display para mostrar o dataframe.

```python
url = 'https://raw.githubusercontent.com/petlove/vagas/master/data/bi-analysis/data-test-analytics.csv'
petlove_df = pd.read_csv(url, sep = ',')
display(petlove_df)
```

# Usei o método .value_counts para calcular a quantidade de assinaturas por status.

# Usei o argumento normalize =True para transformar os números em valores percentuais.

# Usei o map('{:.1%}'.format) com uma funçao para formatar o valor para um percentual com apenas uma casa decimal.

```python
display(petlove_df['status'].value_counts(normalize =True).map('{:.1%}'.format))
```

# Utilizei o método loc para localizar todas as assinaturas com o status 'paused'.

# Armazenei no dataframe assinatura_pausada_df todas as assinaturas pausadas e selecionei as colunas com os dados mais relevantes para a análise das assinaturas pausadas.

# Usei a função display para mostrar o dataframe assinatura_pausada_df.

```python
assinatura_pausada_df = petlove_df.loc[petlove_df['status'] == 'paused', 
['name_hash',
'email_hash',
'birth_date',
'status',
'city',
'state',
'last_date_purchase',
'average_ticket'
]]
display(assinatura_pausada_df)
```

# Utilizei o método loc para localizar todos as assinaturas com o status 'canceled'.

# Armazenei no dataframe assinatura_cancelada_df todas as assinaturas canceladas e selecionei as colunas com os dados mais relevantes para a análise das assinaturas canceladas.¶

# Usei a função display para mostrar o dataframe assinatura_cancelada_df.

```python
assinatura_cancelada_df = petlove_df.loc[petlove_df['status'] == 'canceled', 
['name_hash',
'email_hash',
'birth_date',
'status',
'created_at',
'deleted_at',
'city',
'state',
'average_ticket'
]]
display(assinatura_cancelada_df)
```

# Utilizei o método loc para localizar todos as assinaturas com o status 'active'.

# Armazenei no dataframe assinatura_ativa_df todas as assinaturas ativas e selecionei as colunas com os dados mais relevantes para a análise das assinaturas ativas.

# Usei a função display para mostrar o dataframe assinatura_ativa_df.

```python
assinatura_ativa_df = petlove_df.loc[petlove_df['status'] == 'active',
['name_hash',
'email_hash',
'birth_date',
'status',
'city',
'state',
'average_ticket',
'recency']]
display(assinatura_ativa_df)
```

# Utilizei o método query para consultar e a expressão lógica > para filtrar os dados maiores que 45 na coluna 'recency'.

# Armazenei no dataframe ultima_compra45dias_df o resultado da consulta.

# Usei a função display para mostrar o dataframe ultima_compra45dias_df.

```python
ultima_compra45dias_df = assinatura_ativa_df.query('recency > 45')
display(ultima_compra45dias_df)
```

# Usei a função pd.to_datetime para alterar o data type das colunas 'created_at' e 'deleted_at'.

# Separei e criei duas novas colunas para date e time.

# Usei a função pd.to_datetime para alterar o data type das colunas 'created_day' e 'deleted_day'.

# Usei o método .info() para visualizar as informações do dataframe.

```python
assinatura_cancelada_df['created_at'] = pd.to_datetime(assinatura_cancelada_df['created_at'])
assinatura_cancelada_df['deleted_at'] = pd.to_datetime(assinatura_cancelada_df['deleted_at'])

assinatura_cancelada_df['created_day'] = assinatura_cancelada_df['created_at'].dt.date
assinatura_cancelada_df['created_hour'] = assinatura_cancelada_df['created_at'].dt.time

assinatura_cancelada_df['deleted_day'] = assinatura_cancelada_df['deleted_at'].dt.date
assinatura_cancelada_df['deleted_hour'] = assinatura_cancelada_df['deleted_at'].dt.time

assinatura_cancelada_df['created_day'] = pd.to_datetime(assinatura_cancelada_df['created_day'])
assinatura_cancelada_df['deleted_day'] = pd.to_datetime(assinatura_cancelada_df['deleted_day'])

assinatura_cancelada_df.info()
```

# Fiz um cálculo entre tabelas de data resultando na diferença da quantidade de dias da data de criação pela data de cancelamento.

```python
assinatura_cancelada_df['Tempo de assinatura'] = assinatura_cancelada_df['deleted_day'] - assinatura_cancelada_df['created_day']
display(assinatura_cancelada_df['Tempo de assinatura'])
```
